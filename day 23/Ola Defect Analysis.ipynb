```python
data = "Ride101-Ramesh-Late Arrival-Bangalore, Ride102-Suresh-rude behavior-Hyderabad, Ride103-Mahesh-Late arrival-Bangalore"
```

```python
print(data)
```

```python
# Task 1: Clean the Data (convert to lowercase)
data_lower = data.lower()
print("Cleaned Data:", data_lower)
```

```python
# Task 2: Count Total Complaints
total_complaints = len(data_lower.split(','))
print("Total Complaints:", total_complaints)
```

```python
# Task 3: Count "Late Arrival" Issues
late_arrival_count = data_lower.count('late arrival')
print("Late Arrival Issues:", late_arrival_count)
```

```python
# Task 4: Find Complaints from Bangalore
bangalore_count = data_lower.count('bangalore')
print("Complaints from Bangalore:", bangalore_count)
```

```python
# Task 5: Extract First Driver Name
first_driver = data_lower.split(',')[0].split('-')[1]
print("First Driver Name:", first_driver)
```

```python
# Task 6: Replace Issue Names (Standardization)
data_replaced = data_lower.replace('late arrival', 'late_arrival').replace('rude behavior', 'rude_behavior')
print("Standardized Data:", data_replaced)
```
