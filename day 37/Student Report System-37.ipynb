```python
# 1. Create a function get_marks()
def get_marks():
    marks = []
    for i in range(1, 4):
        mark = int(input(f"Enter marks for subject {i}: "))
        marks.append(mark)
    return marks 
```

```python
# 2. Create a function calculate_total(marks)
def calculate_total(marks):
    return sum(marks)  

# 3. Create a function calculate_average(total)
def calculate_average(total):
    return total / 3  

# 4. Create a function get_grade(avg)
def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"
```

```python
# 5. Main program
def main():
    marks = get_marks()  # Call get_marks()
    total = calculate_total(marks)  
    avg = calculate_average(total)  
    grade = get_grade(avg)
    
    # Display total, average, and grade
    print("\n--- Result ---")
    print("Marks:", marks)
    print("Total:", total)
    print("Average:", round(avg, 2))
    print("Grade:", grade)
```

```python
# Run the program
if __name__ == "__main__":
    main()
```
