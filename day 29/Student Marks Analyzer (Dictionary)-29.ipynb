```python
students = {
 "Asha": {"Math": 78, "Science": 85, "English": 67},
 "Ravi": {"Math": 35, "Science": 40, "English": 50},
 "John": {"Math": 90, "Science": 92, "English": 88}
}
```

```python
## Task 1: Calculate total & average for each student
student_totals = {}
student_averages = {}

for name, subjects in students.items():
    total = sum(subjects.values())
    avg = total / len(subjects)
    student_totals[name] = total
    student_averages[name] = round(avg, 2) 

print("Totals:", student_totals)
print("Averages:", student_averages)
```

```python
##Task 2: Identify top scorer
##Top scorer = student with highest average

top_student = max(student_averages, key=student_averages.get)
print("Top Scorer:", top_student, "with avg", student_averages[top_student])
```

```python
##Task 3: Find students who failed in any subject
##Assuming pass mark = 40, so fail is < 40

failed_students = []

for name, subjects in students.items():
    if any(mark < 40 for mark in subjects.values()):
        failed_students.append(name)

print("Students failed in any subject:", failed_students)
```
