```python
# Sample data
rides = [
    (101, "Ravi", "Bangalore", 250, 4.5),
    (102, "Anjali", "Hyderabad", 180, 4.2),
    (103, "Kiran", "Bangalore", 300, 4.8),
    (104, "Meena", "Chennai", 220, 4.0),
    (105, "Arjun", "Bangalore", 150, 3.9)]
```

```python
# 1. Count total rides
total_rides = len(rides)
print("Total rides:", total_rides)
```

```python
# 2. Extract all rides from Bangalore
bangalore_rides = [ride for ride in rides if ride[2] == "Bangalore"]
print("Bangalore rides:", bangalore_rides)
```

```python
# 3. Find total revenue from Bangalore
bangalore_revenue = sum(ride[3] for ride in rides if ride[2] == "Bangalore")
print("Bangalore revenue:", bangalore_revenue)
```

```python
# 4. Find highest rated driver
highest_rated_driver = max(rides, key=lambda x: x[4])
print("Highest rated driver:", highest_rated_driver)
```

```python
# 5. Average rating per city
city_ratings = {}
for ride in rides:
    city = ride[2]
    rating = ride[4]
    if city in city_ratings:
        city_ratings[city].append(rating)
    else:
        city_ratings[city] = [rating]

for city, ratings in city_ratings.items():
    avg_rating = sum(ratings) / len(ratings)
    print(f"Average rating in {city}: {avg_rating}")
```
