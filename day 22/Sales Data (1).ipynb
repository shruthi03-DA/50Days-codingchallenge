```python
#Task 1: Declare the variables and assign the values as above with appropriate Data Type
order_id = "ORD12345"
customer_name = "Rahul"
product_price = 799.99
quantity = 3
discount = 0.10
is_member = True
```

```python
#Task 2: Identify Data Types & Print the data type of each variable using type().
print(type(order_id))
print(type(customer_name))
print(type(product_price))
print(type(quantity))
print(type(discount))
print(type(is_member))
```

```python
#Task 3: Calculate Total Price • Total price before discount
total_price=product_price * quantity
print("Total Price Before Discount:", total_price)
```

```python
#Task 4: Apply Discount • Total price after discount
discount_amount = total_price * discount
price_after_discount = total_price - discount_amount
print("price_after_discount:" , price_after_discount )
```

```python
#Task 5: Add GST (18%)
#• Calculate GST on the final price • Add it to get the final bill amount 
gst = price_after_discount * 0.18
final_amount = price_after_discount + gst
print("final_amount:", final_amount)
```

```python
#Task 6: Display Output
#Print the final bill in a clean format:
print("----- Final Bill -----")
print("Order ID:", order_id)
print("Customer:", customer_name)
print("Total Price:", round(total_price, 2))
print("Discount Applied:", "Yes" if is_member else "No")
print("Final Amount (including GST):", round(final_amount, 2))
```
